**TASK - 1**
Create a pandas dataframe (DataFrame named as 'df') with numpy random values(4 observations and 4 features)

```python
#imports
import numpy as np
import pandas as pd
```

```python
#first, we generate a random numpy 4X4 array.
numpy_random = np.random.randint(low=0, high=20, size=(4,4))
numpy_random
```

```python
#Now we convert this to a pandas dataframe
df = pd.DataFrame(numpy_random, columns=["feature 1","feature 2","feature 3","feature 4"])
df
```
